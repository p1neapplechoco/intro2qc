```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

# 2 qubit: q0 ở trên, q1 ở dưới
qc = QuantumCircuit(2)

# Đưa qubit 0 về |1>
qc.x(0)

# Chuẩn bị |i> = (|0> + i|1>)/sqrt(2) trên qubit 1
qc.h(1)  # |+> = (|0>+|1>)/sqrt(2)
qc.s(1)  # S: |1> -> i|1>, được |i>

print(qc.draw("text"))
# Lấy statevector của toàn hệ
psi = Statevector.from_instruction(qc)
print(psi)

# Vẽ Bloch cho cả 2 qubit
plot_bloch_multivector(psi)
```

```python
theta = np.pi / 2
phi = np.pi / 6

alpha = np.cos(theta / 2)
beta = np.exp(1j * phi) * np.sin(theta / 2)
psi = Statevector([alpha, beta])

print("Vector trạng thái:")
print(psi)

plot_bloch_multivector(psi)
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import numpy as np

qc = QuantumCircuit(2)

theta = np.pi / 3  # 60 degrees

# qubit 0:  sqrt(3)/2 |0> + 1/2 |1>
qc.ry(theta, 0)

# qubit 1:  sqrt(3)/2 |0> - 1/2 |1>
qc.ry(theta, 1)
qc.z(1)  # đổi dấu biên độ |1>

print(qc.draw("text"))

# kiểm tra statevector
psi = Statevector.from_instruction(qc)
print(psi)

# vẽ Bloch cho cả 2 qubit
plot_bloch_multivector(psi)
```

```python
import numpy as np
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector


q1 = Statevector([np.sqrt(3) / 2, 1 / 2])
q2 = Statevector([np.sqrt(3) / 2, -1 / 2])

psi_3 = q1.tensor(q2)

print("Statevector tổng hệ thống:")
print(psi_3)

plot_bloch_multivector(psi_3)
```

```python
coeffs = [
    3 / 4,  # |00>
    -np.sqrt(3) / 4,  # |01>
    np.sqrt(3) / 4,  # |10>
    -1 / 4,  # |11>
]

psi_check = Statevector(coeffs)
plot_bloch_multivector(psi_check)
```
